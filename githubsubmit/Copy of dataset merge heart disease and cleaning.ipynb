```python
import pandas as pd

#Drop rows with NaN values from both datasets if needed
medical_dataset_filtered = Medicaldataset_google_sheet_version___Medicaldataset.dropna(
    how="any"
)  # Remove rows with NaN values
heart_failure_dataset_filtered = (
    Heart_Failure_Prediction_Dataset___all_sets_combined.dropna(how="any")
)

#check if same number of rows
sample_size = min(
    len(medical_dataset_filtered), len(heart_failure_dataset_filtered)
)  # Adjust as needed

# Randomly sample rows (ensure to sample based on the smaller dataset)
sampled_medical_dataset = medical_dataset_filtered.sample(n=sample_size, random_state=1)
sampled_heart_failure_dataset = heart_failure_dataset_filtered.sample(
    n=sample_size, random_state=1
)

# Reset index to avoid misalignment during concatenation
sampled_medical_dataset = sampled_medical_dataset.reset_index(drop=True)
sampled_heart_failure_dataset = sampled_heart_failure_dataset.reset_index(drop=True)

# Concatenate the two datasets side by side (columns aligned by index)
combined_dataset = pd.concat(
    [sampled_medical_dataset, sampled_heart_failure_dataset], axis=1
)

# Check the final shape and save if needed
print("Number of rows and columns in combined dataset:", combined_dataset.shape)
combined_dataset.to_csv("combineds_medical_heart_failure_data.csv", index=False)
```
